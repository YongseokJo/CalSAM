```python
import numpy as np
from scipy.interpolate import splrep, BSpline
import matplotlib.pyplot as plt
```

```python
a = np.loadtxt("./result/smf_1.dat")
a.shape
```

```python
x, y = a[:,0], a[:,1]
tck = splrep(x, y, s=0)
tck_s = splrep(x, y, s=len(x))
xnew = np.linspace(6,12,15)
plt.plot(x, y, 'k')
plt.plot(xnew, BSpline(*tck)(xnew), 'r')
plt.plot(xnew, BSpline(*tck_s)(xnew), 'b')
plt.yscale("log")
plt.xlim(6,12)
plt.show()
```
